# Summary charts

```python
# hide_input

#dfw = dfw.drop("Age_Group", axis=1)
#dfw = dfw.drop("Client_Gender", axis=1)
dfw = df.drop("Case_AcquisitionInfo", axis=1)
dfw = dfw.drop("Outbreak_Related", axis=1)
dfw = dfw.drop("Outcome1", axis=1)
dfw = dfw.drop("Reporting_PHU_City", axis=1)
dfw

c1 = alt.Chart(dfw).mark_bar().encode(
    alt.X('Age_Group'),
    alt.Y('count()'),
    color='Client_Gender',
)

c2 = alt.Chart(dfw).mark_bar(opacity=.3).encode(
    alt.X('Age_Group'),
    alt.Y('count()', stack=None),
    color='Client_Gender',
)

c3 = alt.Chart(dfw).mark_line().encode(
    alt.X('Age_Group'),
    alt.Y('count()'),
    color='Client_Gender',
)

c1 | c2 | c3
```
